```python
def calculate_haversine_distance(df):
    location_columns = location_subset.columns.tolist()

    lat1 = df[location_columns[0]]
    lon1 = df[location_columns[1]]
    lat2 = df[location_columns[2]]
    lon2 = df[location_columns[3]]

    lon1, lat1, lon2, lat2 = map(
        np.radians, [lon1, lat1, lon2, lat2]
    )

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = (
        np.sin(dlat / 2.0) ** 2
        + np.cos(lat1) * np.cos(lat2)
        * np.sin(dlon / 2.0) ** 2
    )

    c = 2 * np.arcsin(np.sqrt(a))
    distance = 6371 * c

    return (
        df.assign(
            distance=distance
        )
    )
```

```python
# Apply the function to create the new 'distance' column
df = calculate_haversine_distance(df)

# Check the new column along with your coordinates
print(df[['restaurant_latitude', 'restaurant_longitude', 'delivery_latitude', 'delivery_longitude', 'distance']].head())
```
